```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
#Loading the breast cancer dataset directly from sklearn
brstc = datasets.load_breast_cancer()
```

```python
#Extracting data and targets
X_data = brstc.data[:, [7, 23]]
y_data = brstc.target
```

```python
#Splitting both data and targets, into training and testing subsets
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size = 0.2, random_state = 1, stratify = y_data)
```

```python
#Initializing standard scaler
ssc = StandardScaler()
```

```python
#Fitting standard scaler
ssc.fit(X_train)
```

```python
#Applying standard sclaing tranformation on both training and testing subsets
X_train_std = ssc.transform(X_train)
X_test_std = ssc.transform(X_test)
```

```python
#Initialzing Perceptron
ppn = Perceptron(eta0 = 0.001, random_state = 1, max_iter = 2500, n_jobs = -1, early_stopping = True)
```

```python
#Fitting the model
ppn.fit(X_train_std, y_train)
```

```python
#Making predcition using test data, to estimate how well the model classifies data
y_pred = ppn.predict(X_test_std)
```

```python
#Estimating the misses count
print(f"Misses count: {(y_pred != y_test).sum()}")
```

```python
#Estimating the accuracy score
print(f"Accuracy score: {100 * accuracy_score(y_pred, y_test):.2f}%")
```

```python
#Visualizing data
def plot_decision_regions(X, y, classifer, resolution = 0.001):
    markers = ('o', 'v', 'x', 's')
    colors = ('red', 'green', 'blue', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    x1_max = X[:, 0].max() + 1
    x1_min = X[:, 0].min() - 1
    x2_max = X[:, 1].max() + 1
    x2_min = X[:, 1].min() - 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), np.arange(x2_min, x2_max, resolution))
    maper = classifer.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    maper = maper.reshape(xx1.shape)

    plt.contourf(xx1, xx2, maper, cmap = cmap, alpha = 0.7)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x = X[y == cl, 0], y = X[y == cl, 1], marker = markers[idx], alpha = 0.9, label = cl, edgecolor = 'black',
                   c = colors[idx])
```

```python
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined_std = np.hstack((y_train, y_test))
plot_decision_regions(X_combined_std, y_combined_std, ppn)
plt.xlabel('Perimeter 1')
plt.ylabel('Perimeter 2')
plt.legend(loc = 'upper left')
plt.tight_layout()
```
